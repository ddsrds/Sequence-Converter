#2 - File addresses on Google Drive

```python
# Etapa 2: Endereço dos arquivos
caminho_entrada = '/content/drive/MyDrive/scripts/nucprot/entrada.txt'
caminho_saida = '/content/drive/MyDrive/scripts/nucprot/saida.txt'
```

#3 - Open file for reading and editing

```python
# Etapa 3: Tabela de códons → aminoácidos (1 letra)
tabela_codons = {
    'AUG': 'M',  # Met (Início)
    'UUU': 'F', 'UUC': 'F',
    'UUA': 'L', 'UUG': 'L',
    'CUU': 'L', 'CUC': 'L', 'CUA': 'L', 'CUG': 'L',
    'AUU': 'I', 'AUC': 'I', 'AUA': 'I',
    'GUU': 'V', 'GUC': 'V', 'GUA': 'V', 'GUG': 'V',
    'UCU': 'S', 'UCC': 'S', 'UCA': 'S', 'UCG': 'S',
    'AGU': 'S', 'AGC': 'S',
    'CCU': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',
    'ACU': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T',
    'GCU': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',
    'UAU': 'Y', 'UAC': 'Y',
    'CAU': 'H', 'CAC': 'H',
    'CAA': 'Q', 'CAG': 'Q',
    'AAU': 'N', 'AAC': 'N',
    'AAA': 'K', 'AAG': 'K',
    'GAU': 'D', 'GAC': 'D',
    'GAA': 'E', 'GAG': 'E',
    'UGU': 'C', 'UGC': 'C',
    'UGG': 'W',
    'CGU': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R',
    'AGA': 'R', 'AGG': 'R',
    'GGU': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G',
    'UAA': 'Stop', 'UAG': 'Stop', 'UGA': 'Stop'
}

# Função para traduzir RNA → sequência de aminoácidos (1 letra)
def traduzir_rna(rna):
    proteina = ''
    rna = rna.upper().replace(" ", "").replace("\n", "")
    for i in range(0, len(rna), 3):
        codon = rna[i:i+3]
        if len(codon) != 3:
            break
        aminoacido = tabela_codons.get(codon, '')
        if aminoacido == 'Stop':
            break
        proteina += aminoacido
    return proteina

# Etapa 4: Processar entrada e montar saída no estilo FASTA
linhas_saida = []
with open(caminho_entrada, 'r') as f:
    titulo = ''
    sequencia = ''
    for linha in f:
        linha = linha.strip()
        if linha.startswith('>'):
            if titulo and sequencia:
                proteina = traduzir_rna(sequencia)
                linhas_saida.append(f'>{titulo}')
                linhas_saida.append(proteina)
            titulo = linha[1:]  # remove o >
            sequencia = ''
        else:
            sequencia += linha
    # Finaliza última sequência
    if titulo and sequencia:
        proteina = traduzir_rna(sequencia)
        linhas_saida.append(f'>{titulo}')
        linhas_saida.append(proteina)

# Etapa 5: Escrever saída no arquivo
with open(caminho_saida, 'w') as f:
    for linha in linhas_saida:
        f.write(linha + '\n')

# Etapa 6: Exibir no console
print("Arquivo de saída gerado com sucesso:")
for linha in linhas_saida:
    print(linha)
```
